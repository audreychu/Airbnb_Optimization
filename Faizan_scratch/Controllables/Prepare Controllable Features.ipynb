```python
#Basic EDA (review missing data, get means, etc.)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
def getDfSummary(input_data):
    output_data = pd.DataFrame({'number_nan': input_data.isna().sum()})
    output_data['number_distinct'] = input_data.nunique()
    get_descriptions = input_data.describe().transpose()[['mean', 'max', 'min', 'std', '25%', '50%', '75%']]
    output_data = pd.concat([output_data, get_descriptions], axis=1) 
    
    return output_data

getDfSummary(df_cleansed)
```
